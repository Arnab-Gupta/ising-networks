```
import networkx as nx
from isingnetworks import IsingModel
import numpy as np
g = nx.erdos_renyi_graph(20, 0.4)
```

```
model = IsingModel(g)
```

```
J = 1.0
temperature = np.arange(1, 10, 0.1)
initial_state = 1

iterations = 10000
model.viz(J, temperature, iterations, initial_state)
```
